```python
import pandas as pd
from sklearn.model_selection import train_test_split

# 讀取CSV檔案
file_path = './dataset/data.csv'
df = pd.read_csv(file_path)

# 檢查缺失值
null_values = df.isnull().sum()

# 如果有缺失值，可以使用 dropna() 丟棄包含缺失值的行
if null_values.any():
    print("Data contains missing values. Dropping rows with missing values.")
    df = df.dropna()

# 分割資料集
train_data, temp_data = train_test_split(df, test_size=0.4, random_state=42)
valid_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)

# 顯示每個部分的大小
print(f'Train set size: {len(train_data)}')
print(f'Validation set size: {len(valid_data)}')
print(f'Test set size: {len(test_data)}')

# 將分割後的資料存為新的CSV檔案
train_data.to_csv('./dataset/train_set.csv', index=False)
valid_data.to_csv('./dataset/valid_set.csv', index=False)
test_data.to_csv('./dataset/test_set.csv', index=False)
```

```python
from textattack.augmentation import (
    EasyDataAugmenter,
    WordNetAugmenter,
    DeletionAugmenter,
    EmbeddingAugmenter,
    CharSwapAugmenter,
    CheckListAugmenter,
    BackTranslationAugmenter,
)

# Create augmenters
easy_data_augmenter = EasyDataAugmenter()
word_net_augmenter = WordNetAugmenter()
deletion_augmenter = DeletionAugmenter()
embedding_augmenter = EmbeddingAugmenter()
char_swap_augmenter = CharSwapAugmenter()
checklist_augmenter = CheckListAugmenter()
back_translation_augmenter = BackTranslationAugmenter()

# Input string
s = "炒鳝拆鳝丝炒之，略焦，如炒肉鸡之法，不可用水"

# Augment each function and save the output to the same file
output_file_path = "all_augmented_outputs.txt"

with open(output_file_path, "w", encoding="utf-8") as output_file:
    output_file.write(f"Original Text:\n{s}\n")

    augmenters = {
        "EasyDataAugmenter": easy_data_augmenter,
        "WordNetAugmenter": word_net_augmenter,
        "DeletionAugmenter": deletion_augmenter,
        "EmbeddingAugmenter": embedding_augmenter,
        "CharSwapAugmenter": char_swap_augmenter,
        "CheckListAugmenter": checklist_augmenter,
        "BackTranslationAugmenter": back_translation_augmenter,
    }

    for augmenter_name, augmenter in augmenters.items():
        augmented_string = augmenter.augment(s)
        output_file.write(f"\n{augmenter_name}:\n{augmented_string}\n")

print(f"All augmented outputs saved to {output_file_path}")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from transformers import T5ForSequenceClassification, T5Tokenizer
from sklearn.metrics import accuracy_score
import pandas as pd

# Load the datasets
train_df = pd.read_csv('./dataset/train_set.csv')
valid_df = pd.read_csv('./dataset/valid_set.csv')
test_df = pd.read_csv('./dataset/test_set.csv')

class CustomDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = int(self.labels[idx])

        inputs = self.tokenizer(
            text,
            return_tensors='pt',
            max_length=self.max_length,
            truncation=True,
            padding='max_length',
        )

        return {
            'input_ids': inputs['input_ids'].squeeze(),
            'attention_mask': inputs['attention_mask'].squeeze(),
            'labels': torch.tensor(label),
        }

# Set up the T5 model and tokenizer
model_name = 't5-base'
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForSequenceClassification.from_pretrained(model_name, num_labels=len(train_df['label'].unique()))

# Prepare the datasets
train_dataset = CustomDataset(train_df['train_content'], train_df['label'], tokenizer, max_length=128)
valid_dataset = CustomDataset(valid_df['train_content'], valid_df['label'], tokenizer, max_length=128)
test_dataset = CustomDataset(test_df['train_content'], test_df['label'], tokenizer, max_length=128)

# Set up the data loaders
train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=8, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False)

# Set up the optimizer and loss function
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
criterion = torch.nn.CrossEntropyLoss()

# Training loop
num_epochs = 3
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

for epoch in range(num_epochs):
    model.train()
    total_loss = 0.0

    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        total_loss += loss.item()

        loss.backward()
        optimizer.step()

    average_loss = total_loss / len(train_loader)
    print(f'Epoch {epoch + 1}/{num_epochs}, Loss: {average_loss:.4f}')

# Validation loop
model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for batch in valid_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Calculate accuracy on the validation set
accuracy = accuracy_score(all_labels, all_preds)
print(f'Validation Accuracy: {accuracy:.4f}')

# Test loop
model.eval()
all_test_preds = []
all_test_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds = torch.argmax(logits, dim=1)

        all_test_preds.extend(preds.cpu().numpy())
        all_test_labels.extend(labels.cpu().numpy())

# Calculate accuracy on the test set
test_accuracy = accuracy_score(all_test_labels, all_test_preds)
print(f'Test Accuracy: {test_accuracy:.4f}')
```
